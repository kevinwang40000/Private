```python
def multi_plot(df, title, addAll = True):
    fig = go.Figure()
    for column in df.columns.to_list():
        fig.add_trace(
            go.Scatter(
                x = df.index,
                y = df[column],
                name = column
            )
        )
    button_all = dict(label = 'All',
                      method = 'update',
                      args = [{'visible':         
    df.columns.isin(df.columns),
                               'title': 'All',
                               'showlegend':True}])

    
    def create_layout_button(column):
        return dict(label = column,
                    method = 'update',
                    args = [{'visible': df.columns.isin([column]),
                             'title': column,
                             'showlegend': True}])
    fig.update_layout(
        updatemenus=[go.layout.Updatemenu(
            active = 0,
            buttons = ([button_all] * addAll) +  list(df.columns.map(lambda column: create_layout_button(column)))
            )
        ],
         #yaxis_type="log"       
    )
    # Update remaining layout properties
    fig.update_layout(
        title_text=title,
        height=800
        
    )
   
    
    fig.show()
```

```python
for i in range(1,8):
    col_temp = data.columns[i]
    name_temp = data.columns[i]+'分大区时间序列'
    data_temp = data.iloc[:,[0,i]].pivot(columns='日期',values=col_temp).T
    multi_plot(data_temp, title=name_temp)
```
